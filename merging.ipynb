```python
import pandas as pd

# customer dataframe
df_customer = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'Name': ['Inam', 'sakib', 'Osama']
})

# order dataframe
df_order = pd.DataFrame({
    'CustomerID': [1, 2, 4],
    'OrderAmount': [250, 450, 360]
})

df_merged = pd.merge(df_customer, df_order, on="CustomerID", how="right")

print("right Join")
print(df_merged)  

# cartcian join
# 1df = m rows
# 2df = n rows 
# m*n rows
```
