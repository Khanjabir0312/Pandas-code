```python
# interpolations
#preserve data interrity
#smooth trands
#avoid data loss
#inteerpolate()

import pandas as pd

data ={
  
  "Name":['Inam','Ahmad','Osama','Jabir','Sakib','Mokit','Rehan','Tanvir'],
  "Age":[21,None,23,24,18,14,12,8],
  "Salary":[100000,None,300000,500000,6000000,70000000,4000000,900000],
  "Proformance Score":[90,None,50,60,40,80,70,66]
}

df= pd.DataFrame(data)
print(df)

#liner,polynomial,time(ye sab method bahot imporant hai aur bhi methods hai apko saikhna hoga )

# axis=0 means rows 
# axis=1 means column

df.interpolate(method='linear', axis=0, inplace=True)

```

```python
# linear interpolation

import pandas as pd

data={
    'Time':[1,2,3,4,5],
    'Value':[10,None,30,None,50]
}
df=pd.DataFrame(data)
print('Before interpolation')
print(df)

df['Value'] = df['Value'].interpolate(method="linear")

print('After Interpolation')
print(df)

#ye bahot important hai  Math is mai kese kaam kar raha hai mere ko ye talsh karna hai
# 1-timmer series data
# 2-numeric data with trends
# 3-avoid dropping rows
```
