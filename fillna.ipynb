```python
#fillna()
#fillna(value, inplace=Ture)
import pandas as pd

data ={
  
  "Name":['Inam',None,'Osama','Jabir','Sakib','Mokit','Rehan','Tanvir'],
  "Age":[21,None,23,24,18,14,12,8],
  "Salary":[100000,None,300000,500000,6000000,70000000,4000000,900000],
  "Proformance Score":[90,None,50,60,40,80,70,66]
}

df= pd.DataFrame(data)
print(df)

# df.fillna(0,inplace=True)
# Name column me string fill karna
df['Name'].fillna("Ahamed", inplace=True)

# Numeric columns me mean fill karna
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
df['Proformance Score'].fillna(df['Proformance Score'].mean(),inplace=True)


print(df)
```
