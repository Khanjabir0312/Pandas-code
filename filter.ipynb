```python
#how to filter rows in pandas

import pandas as pd

data={
    "Name":['Inam','Osama','Jabir','Sakib','Mokit','Rehan','Tanvir'],
  "Age":[21,23,24,18,14,12,8],
  "Salary":[100000,300000,500000,6000000,70000000,4000000,900000],
  "Proformance Score":[90,50,60,40,80,70,66]

}

df=pd.DataFrame(data)
#single condition ke liye hai ye 
high_salary = df[df['Salary']>500000]
print('Employee with salary>50000')
print(high_salary)

#multiple condition ke liye 
# filtering rows salary>50000l and age>30

filtered=df[(df['Age']>8) & (df['Salary']>500000)]
print(f'Employee list Age >30 + Salary > 50000')
print(filtered)

# using OR condition 
filtered_or = df[(df["Age"] > 8) | (df["Proformance Score"] > 40)]
print("Employee Age > 8 OR Performance Score > 40")
print(filtered_or)
```
